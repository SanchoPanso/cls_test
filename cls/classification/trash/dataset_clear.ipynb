```python
import json
import pandas as pd

path_json = '/home/timssh/ML/TAGGING/DATA/datasets/tits_size.json'
list_path = '/home/timssh/ML/TAGGING/DATA/datasets/bad_lists/tits_size.txt'

with open(path_json) as f:
    json_ = json.load(f)

json_.keys()
```

```python
df = pd.read_json(json_["data"])
df
```

```python
cols = df.columns[1:]
all_ = 0
for col in cols:
    ret = df[col][df[col]==1].sum()
    print(ret)
    all_ += ret
print(all_)
```

```python
with open(list_path) as f:
    bad_list = f.read()
bad_list = list(bad_list.split('\n'))

df = df[~df['path'].isin(bad_list)]
df = df.reset_index(drop=True)
df
```

```python
json_['data'] = df.to_json()
with open(path_json, "w") as f:
    f.write(json.dumps(json_))
```
